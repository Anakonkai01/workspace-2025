```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# sklearn libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```

```python
# load data to dataframe
iris_raw = load_iris()
iris_pd = pd.DataFrame(iris_raw.data, columns=iris_raw.feature_names)
iris_pd.head()
```

```python
dict(enumerate(iris_raw.target_names))
```

```python
iris_pd['species'] = pd.Series(iris_raw.target)
iris_pd['species'] = iris_pd['species'].map(dict(enumerate(iris_raw.target_names)))
iris_pd.head()
```

```python
iris_pd.hist()
```

```python
sns.pairplot(iris_pd, hue='species', markers=['o', 's', 'D'])
plt.show()
```
